## Q1.What is a random variable in probability theory?
ANS: A random variable is a function that assigns numerical values to the outcomes of a random experiment. It can be discrete(finite/countable outcomes) or continuous(infinite/unbounded outcomes).
## Q2.What are the types of random variables?
ANS:**Discrete Random Variable:** Takes countable values (e.g., number of heads in a coin toss).
**Continuous Random Variable:** Takes any value within a range (e.g., height of individuals).
## Q3.What is the difference between discrete and continuous distributions?
**Discrete Distribution:** Probability is assigned to distinct values (e.g., Binomial, Poisson).
**Continuous Distribution:** Probability is spread over an interval (e.g., Normal, Exponential).
## Q4.What are probability distribution functions (PDF)?
ANS:A Probability Distribution Function (PDF) describes the likelihood of a random variable taking on a particular value.
For discrete variables, it's called the Probability Mass Function (PMF), which assigns probabilities to specific values.
For continuous variables, the Probability Density Function (PDF) represents the probability density at different points (area under the curve gives probability).
## Q5.How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
ANS: The Cumulative Distribution Function (CDF) gives the probability that a random variable takes a value less than or equal to a given number.
Unlike PDF, which shows individual probabilities, CDF accumulates probabilities from the start of the distribution..
## Q6.What is a discrete uniform distribution?
ANS: A discrete uniform distribution has equal probability for all possible outcomes.
Example: Rolling a fair six-sided die, where each outcome (1 to 6) has a probability of 1/6.
## Q7.What are the key properties of a Bernoulli distribution?
ANS: Two possible outcomes: Success (1) or Failure (0).
Probability of success: p, probability of failure: 1 - p.
Mean = p, Variance = p(1 - p).Example: Flipping a coin (heads = 1, tails = 0).
## Q8.What is the binomial distribution, and how is it used in probability?
ANS: Models the number of successes in n independent Bernoulli trials.
Example: Counting the number of heads in 10 coin flips.
## Q9.What is the Poisson distribution, and where is it applied?
ANS:Models the number of events occurring in a fixed interval of time/space, assuming events happen independently and at a constant average rate (λ).
Example: Number of calls at a call center per hour.
## Q10.What is a continuous uniform distribution?
A distribution where all values in an interval have equal probability.
Example: Randomly selecting a number between 0 and 1.
## Q11.What are the characteristics of a normal distribution?
Bell-shaped, symmetric around the mean.
Mean = Median = Mode.
Fully defined by mean (μ) and standard deviation (σ).
Example: Human height distribution.
## Q12.What is the standard normal distribution, and why is it important?
ANS:A normal distribution with mean = 0 and standard deviation = 1.
Used in Z-score calculations to standardize values across distributions.
## Q13.What is the Central Limit Theorem (CLT), and why is it critical in statistics?
ANS:The CLT states that the sampling distribution of the sample mean approaches a normal distribution, even if the population isn't normal, as the sample size increases.
It's crucial because it allows us to use normal distribution-based inference in real-world problems.
## Q14.How does the Central Limit Theorem relate to the normal distribution?
ANS:It explains why many real-world data distributions appear normal, even if the original data isn’t.
## Q15.What is the application of Z statistics in hypothesis testing?
ANS:Z-tests compare sample means to a known population mean to determine if the difference is statistically significant.

## Q18.What are point estimates and interval estimates in statistics?
ANS:Point Estimate: A single value used to estimate a population parameter (e.g., sample mean).
Interval Estimate: A range of values likely to contain the population parameter (e.g., confidence interval).
## Q19.What is the significance of confidence intervals in statistical analysis?
ANS:They provide a range where the true population parameter is likely to be found, with a given confidence level (e.g., 95%).
## Q20.What is the relationship between a Z-score and a confidence interval?
ANS:Z-scores determine the margin of error in confidence intervals.
Example: A 95% confidence interval uses a Z-score of 1.96.
## Q21.How are Z-scores used to compare different distributions?
ANS:Since Z-scores standardize values, they allow comparisons across different datasets with different units or scales.

```python
# 9.Write a Python function to calculate Z-scores from a dataset and plot them
def z_scores(data):
    mean = np.mean(data)
    std_dev = np.std(data)
    return (data - mean) / std_dev
data = np.random.normal(50, 10, 1000)
z_scores_data = z_scores(data)
plt.hist(z_scores_data, bins=30, density=True, alpha=0.6, color='orange')
plt.xlabel("Z-score")
plt.ylabel("Probability Density")
plt.title("Z-score Distribution")
plt.show()
```

```python
#10.Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.
def clt_th(sample_size, num_samples):
    means = [np.mean(np.random.exponential(scale=2, size=sample_size)) for _ in range(num_samples)]
    plt.hist(means, bins=30, density=True, alpha=0.6, color='brown')
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.title("Central Limit Theorem")
    plt.show()

clt_th(30, 1000)
```

```python
#11.Simulate multiple samples from a normal distribution and verify the Central Limit Theorem
def clt_normal_(sample_size, num_samples, mu=50, sigma=15):
    sample_means = [np.mean(np.random.normal(mu, sigma, sample_size)) for _ in range(num_samples)]
    plt.hist(sample_means, bins=30, density=True, alpha=0.6, color='r')
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.title(f"CLT Verification (Normal Dist, n={sample_size})")

    # Overlay normal distribution curve
    x = np.linspace(min(sample_means), max(sample_means), 100)
    normal_curve = stats.norm.pdf(x, mu, sigma/np.sqrt(sample_size))
    plt.plot(x, normal_curve, 'y', label="Expected Normal Curve")

    plt.legend()
    plt.show()

clt_normal_(60, 1000)
```

```python
#12.Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1)
def plot_standard_normal():
    x = np.linspace(-4, 4, 1000)  # Generate values from -4 to 4
    y = stats.norm.pdf(x, 0, 1)   # Calculate PDF for mean=0, std=1

    plt.plot(x, y, color='blue', label='Standard Normal Distribution')
    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.title("Standard Normal Distribution (Mean=0, Std=1)")
    plt.legend()
    plt.show()

plot_standard_normal()
```

```python
#14.Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution
def calculate_z_score(value, data):
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)
    return (value - mean) / std_dev
data = np.random.normal(50, 10, 1000)
data_point = 90
z_score = calculate_z_score(data_point, data)
print(f"Z-score for {data_point}: {z_score:.2f}")

# Compare with standard normal distribution
probability = stats.norm.cdf(z_score)
print(f"Probability of a value ≤ {data_point} in standard normal distribution: {probability:.4f}")
```

```python
#15.Implement hypothesis testing using Z-statistics for a sample dataset
sample_data = [52, 48, 55, 49, 50, 47, 53, 51, 54, 50]
sample_mean = np.mean(sample_data)
sample_size = len(sample_data)
pop_mean = 50
pop_std = 5

z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))
alpha = 0.05
if p_value < alpha:
    result = "Reject the null hypothesis (significant difference)"
else:
    result = "Fail to reject the null hypothesis (no significant difference)"
print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.4f}")
print(f"Conclusion: {result}")
```

```python
#16.Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean
data = np.random.normal(loc=50, scale=10, size=30)
mean = np.mean(data)
std_error = stats.sem(data)
confidence = 0.95
lower, upper = stats.norm.interval(confidence, loc=mean, scale=std_error)
# Print results
print(f"Sample Mean: {mean:.2f}")
print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")
```
